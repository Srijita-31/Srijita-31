```python
import pandas as pd
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, HRFlowable, PageBreak
from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from datetime import datetime



file_path = "/content/gdrive/MyDrive/PythonProg_Data/Medical-Test-PlasmaTech-Diagnostics-data.xls"
df = pd.read_excel(file_path)
start_date = df.iloc[1,1]
end_date = df.iloc[1,3]
hospital_clinic = df.iloc[4,0]
colNames=[df.iloc[3,0], df.iloc[3,1],df.iloc[3,2],df.iloc[3,3],df.iloc[3,4],df.iloc[3,5], df.iloc[3,6],df.iloc[3,7],df.iloc[3,8],df.iloc[3,9],df.iloc[3,10]]



DF = pd.read_excel(file_path,skiprows=5, header=None, names=colNames)
DF.head()
DF['Date'] = pd.to_datetime(DF['Date'], format='%d-%m-%Y %H:%M')


DF['Date'] = DF['Date'].dt.date


start_date = datetime.strptime(start_date, '%d-%m-%Y').date()
end_date = datetime.strptime(end_date, '%d-%m-%Y').date()


filtered_df = DF[(DF['Date'] >= start_date) & (DF['Date'] <= end_date)]


grouped_data_by_visit = filtered_df.groupby('Patient').agg({'Patient': 'count'}).rename(
    columns={'Patient': 'visit_count'}).reset_index()

#Sorting the top 20 patients based on visit to hospital
top_20_by_visits = grouped_data_by_visit.sort_values(by='visit_count', ascending=False).head(20)


#Sorting the bottom 20 patients based on visit to hospital
bottom_20_by_visits = grouped_data_by_visit.sort_values(by='visit_count', ascending=False).tail(20)


#Grouping by 'Patient' and aggregate both the 'Total Price' and count of visits in hospital
grouped_data = filtered_df.groupby('Patient').agg({'Patient': 'count', 'Total Price': 'sum'}).rename(
    columns={'Patient': 'visit_count'}).reset_index()

#Sorting the DataFrame by 'bill_amount' and 'visit_count' in descending order
sort_by_Totalbill = grouped_data.sort_values(by='Total Price', ascending=False)

#Getting the top 20 patients based on bill amount
top_20_by_bill = sort_by_Totalbill.head(20)

#Getting the bottom 20 patients based on bill amount
bottom_20_by_bill = sort_by_Totalbill.tail(50)



pdf_filename = 'Hospital_Data_Analysis.pdf'

page_width, page_height = letter

# Define margin sizes
left_margin = 1 * inch
right_margin = 1 * inch
top_margin = 1 * inch
bottom_margin = 1 * inch

# Creating a document template with adjusted margins
doc = SimpleDocTemplate(pdf_filename, pagesize=(page_width, page_height), leftMargin=left_margin,
                            rightMargin=right_margin, topMargin=top_margin, bottomMargin=bottom_margin)


# Creating a list to hold the tables
elements = []

style = TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.darkblue),
    ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
    ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
    ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
    ('FONTSIZE', (0, 0), (-1, 0), 16),
    ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
    ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
    ('FONTSIZE', (0, 1), (-1, -1), 14),
    ])

styles = getSampleStyleSheet()


footer_style = ParagraphStyle(
    'Footer',
    parent=styles['Normal'],
    spaceBefore=1,
    fontSize=12,
    textColor=colors.blue,
    alignment=1,
)


def add_table_to_page(doc,table_data, page_title, column_headings):



    table_data_with_serial = [['Serial'] + column_headings] + [[str(i + 1)] + row for i, row in enumerate(table_data)]
    elements.append(Paragraph(page_title, styles['Title']))

    table_data.insert(0, column_headings)
    table = Table(table_data_with_serial, repeatRows=1, colWidths=[80] + [200, 100, 100])

    table.setStyle(style)
    elements.append(table)


    elements.append(Spacer(1, 144))


    elements.append(HRFlowable(width="100%", thickness=2, color="green"))

    # Adding footer
    footer_text = "Develped by KRISKO SOFTWARE PRIVATE LIMITED"
    footer_paragraph = Paragraph(footer_text, footer_style)
    elements.append(Spacer(1, 1))  # Adding some space before the footer
    elements.append(footer_paragraph)


    elements.append(PageBreak())

#Adding top 20 patients based on number of visits to the first page
add_table_to_page(doc, top_20_by_visits.values.tolist(), 'Top 20 patient names based on number of visits to %s from  %s to %s'
                   % (hospital_clinic, start_date, end_date), ['Patient Name', 'No. of Visit'])

#Adding bottom 20 patients based on number of visits to the second page
add_table_to_page(doc, bottom_20_by_visits.values.tolist(), 'Bottom 20 patient names based on number of visits to %s from %s to %s'
                   % (hospital_clinic, start_date,end_date), ['Patient Name', 'No. of Visit'])

#Adding top 20 patients based on bill amount to the third page
add_table_to_page(doc, top_20_by_bill.values.tolist(),
                  'Top 20 patient names based on total bill amount with number of visits to %s from %s to %s'
                   % (hospital_clinic, start_date,end_date), ['Patient Name', 'No. of Visit','Bill Amount'])

#Adding bottom 20 patients based on bill amount to the fourth page
add_table_to_page(doc, bottom_20_by_bill.values.tolist(),
                  'Bottom 20 patient names based on total bill amount with number of visits to %s from %s to %s'
                   % (hospital_clinic, start_date,end_date), ['Patient Name', 'No. of Visit','Bill Amount'])

#Building the PDF
doc.build(elements)

print(f'PDF saved successfully: {pdf_filename}')
```
